# Decision Tree using CART

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### I have made the train and test split as 70,30 i.e 70% for training and 30% for testing. This helps in avoiding overfitting as the dataset is small.

```python
# Create a decision tree classifier
dt_classifier = DecisionTreeClassifier(criterion='gini', max_depth=2, min_samples_leaf=2, random_state=42)
```

```python
# Train the classifier on the training data
dt_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the training data
y_train_pred = dt_classifier.predict(X_train)
```

```python
# Make predictions on the test data
y_test_pred = dt_classifier.predict(X_test)
```

```python
# Calculate accuracy on the training set
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.2f}")

# Calculate accuracy on the test set
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Testing Accuracy: {test_accuracy:.2f}")
```

```python
# Calculate and print the classification report
classification_rep = classification_report(y_test, y_test_pred, target_names=iris.target_names)
print("Classification Report:\n", classification_rep)
```

```python
# Plot the tree 
class_names_list = list(iris.target_names)
plt.figure(figsize=(12, 6))
plot_tree(dt_classifier, filled=True, feature_names=iris.feature_names, class_names=class_names_list)
plt.show()
```

### I have used CART for classification and used Gini as criteria for creating the decision tree

### I have also pruned the dataset with minimizing the depth of the tree and increased the minimum samples per leaf.
### This has helped in avoiding overfitting and giving acceptable accuracy for both the training and testing.
